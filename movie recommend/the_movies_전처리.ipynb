```python
import numpy as np
import pandas as pd
```

```python
movie_df = pd.read_csv('./The Movies Dataset/movies_metadata.csv', low_memory=False)
print(len(movie_df))
movie_df.head().transpose()
```

```python
movie_df[(~movie_df['imdb_id'].isna()) & (~movie_df['imdb_id'].astype(str).str.startswith('tt'))]
```

```python
mini_movie_df = movie_df[['imdb_id', 'title', 'genres', 'original_language', 'release_date']].copy()
mini_movie_df = mini_movie_df[mini_movie_df['imdb_id'].str.startswith('tt', na=False)]
mini_movie_df['imdbId'] = mini_movie_df['imdb_id'].map(lambda i: int(i.lstrip('tt')))
mini_movie_df = mini_movie_df[~mini_movie_df['title'].isna()]
mini_movie_df['genres'] = mini_movie_df['genres'].map(lambda gs: ', '.join([g['name'] for g in eval(gs)]))
mini_movie_df = mini_movie_df.drop(columns=['imdb_id'])
print(len(mini_movie_df))
mini_movie_df.head()
```

```python
mini_movie_df.to_csv('./The Movies Dataset/processed_movies.csv', index=False)
```

```python
link_df = pd.read_csv('./The Movies Dataset/links.csv')
print(len(link_df))
rating_df = pd.read_csv('./The Movies Dataset/ratings.csv')
print(len(rating_df))
rating_df.head()
```

```python
rating_df[~rating_df['movieId'].isin(link_df['movieId'])]
```

```python
imdb_rating_df = (
    rating_df
    .join(
        link_df.set_index('movieId')[['imdbId']], 
        how='inner', 
        on='movieId',
    )
    .drop(columns=['movieId'])
)
print(len(imdb_rating_df))
imdb_rating_df.head()
```

```python
imdb_rating_df.to_csv('./The Movies Dataset/processed_ratings.csv', index=False)
```
